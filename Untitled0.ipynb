```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing import image as keras_image_utils
```

```python
# 图像设置...
# BATCH_SIZE = 32

# 数据集路径 (请根据你的实际路径修改)
base_dir = '/content/drive/MyDrive/智慧辨識檢測/10 27'  # <--- 请确保已修改为你的数据集根目录
train_dir = os.path.join(base_dir, 'train')         # <--- 确保这一行被执行
validation_dir = os.path.join(base_dir, 'validation')
```

```python
# VGG16 模型在 ImageNet 上训练时，使用了特定的预处理（例如，将像素值缩放到 VGG16 的要求）
# Keras 的 applications 模块提供了对应的预处理函数
from tensorflow.keras.applications.vgg16 import preprocess_input

# 训练集生成器（进行数据增强和预处理）
train_datagen = ImageDataGenerator(
    rotation_range=20,        # 随机旋转角度
    width_shift_range=0.2,    # 随机水平平移
    height_shift_range=0.2,   # 随机垂直平移
    shear_range=0.2,          # 剪切变换
    zoom_range=0.2,           # 随机缩放
    horizontal_flip=True,     # 随机水平翻转
    fill_mode='nearest',
    preprocessing_function=preprocess_input  # VGG16 预处理
)

# 验证集生成器（只进行预处理，不进行增强）
validation_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input
)

# 从目录生成数据流
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary'  # 二分类问题
)

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary'
)
```

```python
# 加载 VGG16 基础模型
# weights='imagenet': 使用在 ImageNet 上预训练的权重
# include_top=False: 不包含 VGG16 顶部的全连接分类层
# input_shape: 指定输入图像的尺寸
vgg16_base = VGG16(
    weights='imagenet',
    include_top=False,
    input_shape=(IMG_SIZE, IMG_SIZE, 3)
)

# 冻结 VGG16 卷积层，防止训练过程中权重被更新
vgg16_base.trainable = False

# 构建最终模型
model = Sequential([
    vgg16_base,           # 添加 VGG16 基础模型
    Flatten(),            # 将 3D 输出展平为 1D
    Dense(256, activation='relu'),  # 新的全连接层
    Dropout(0.5),         # 0.5 的 dropout 防止过拟合
    Dense(1, activation='sigmoid')  # 最终输出层，二分类使用 sigmoid
])

# 编译模型
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
    loss='binary_crossentropy',   # 二分类损失函数
    metrics=['accuracy']
)

# 打印模型摘要
model.summary()
```

```python
# 定义训练的 epoch 数量
EPOCHS = 10

# 开始训练
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // BATCH_SIZE, # 每次 epoch 的步数
    epochs=EPOCHS,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // BATCH_SIZE
)
```

```python
# 绘制训练结果
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

```python
def predict_and_show(img_path, model):
    """
    加载图片，进行预处理，使用训练好的模型进行预测，并显示结果。
    """
    try:
        # 1. 加载和调整图片大小 (VGG16需要 224x224)
        img = keras_image_utils.load_img(
            img_path,
            target_size=(IMG_SIZE, IMG_SIZE)
        )

        # 2. 转换成 NumPy 数组
        img_array = keras_image_utils.img_to_array(img)

        # 3. 增加一个维度 (Batch size: (1, 224, 224, 3))
        img_array = np.expand_dims(img_array, axis=0)

        # 4. VGG16 特定的预处理
        processed_image = preprocess_input(img_array)

        # 5. 进行预测
        prediction = model.predict(processed_image)

        # 6. 解释预测结果
        # 由于我们使用 Sigmoid (二分类)，输出是一个 0 到 1 的概率值。
        # 训练生成器中 'cat' 是 0, 'dog' 是 1。
        # 如果预测值大于 0.5，我们判定为 Dog。

        probability = prediction[0][0]

        if probability > 0.5:
            label = f"Dog ({probability:.2f})"
        else:
            label = f"Cat ({1 - probability:.2f})"

        # 7. 显示图片和结果
        plt.figure()
        # 显示原始图片 (非预处理后的数组)
        plt.imshow(img)
        plt.title(f"Prediction: {label}")
        plt.show()

    except Exception as e:
        print(f"处理图片时发生错误: {e}")
        print("请检查图片路径是否正确或文件是否损坏。")


# 假设模型的训练过程已经完成，且模型变量名为 'model'
# IMG_SIZE 和 preprocess_input 也已在之前的单元格中定义。
```
